```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
tweets_df=pd.read_csv('dataset.csv')
```

```python
tweets_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the CountVectorizer
vectorizer = CountVectorizer(ngram_range=(1, 2))

# Fit and transform the tweet text
X = vectorizer.fit_transform(tweets_df['Processed_Tweets'])

# Get the feature names
feature_names = vectorizer.get_feature_names()

# Convert the features to a pandas dataframe
features_df = pd.DataFrame(X.toarray(), columns=feature_names)
```
