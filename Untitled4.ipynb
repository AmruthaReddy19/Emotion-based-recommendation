```python
import pandas as pd
import re
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords,words
nltk.download('stopwords')
```

```python
tweets_df=pd.read_csv('final_data.csv')
```

```python
tweets_df
```

```python
tweets_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
tweets_df.drop('Unnamed: 0.1',axis=1,inplace=True)
```

```python
tweets_df.head()
```

```python
stop_words = list(stopwords.words('english'))
```

```python
def preprocessTweets(tweet):
  tweet=tweet.lower()
  #Remove Urls
  tweet = re.sub(r"http\S+|www\S+|https\S+", '', tweet)
  #remove @reference and # from tweet
  tweet = re.sub(r'\@\w+|\#|\d+','',tweet)
  #remove stopword
  tweet_token =word_tokenize(tweet) #Convert string to token
  filtered_words =  [w for w in tweet_token if w not in stop_words]
  

  

  return "".join(filtered_words)
```

```python
tweets_df['Processed_Tweets']= tweets_df['Tweet'].apply(preprocessTweets)
```
