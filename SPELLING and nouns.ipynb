```python
from textblob import TextBlob
```

```python
import re

# Clean the tweet text
df['clean_text'] = df['Tweets_Sentiment'].apply(lambda x: ' '.join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", x).split()))

# Calculate polarity and subjectivity using TextBlob
df['polarity'] = df['Tweets_Sentiment'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['Tweets_Sentiment'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Print the results
print(df[['Tweets_Sentiment', 'polarity', 'subjectivity']])
```

```python
import nltk
```

```python
df['tokens'] = df['clean_text'].apply(nltk.word_tokenize)
```

```python
df['pos_tags'] = df['tokens'].apply(nltk.pos_tag)

# Extract the nouns
df['nouns'] = df['pos_tags'].apply(lambda x: [word[0] for word in x if word[1]=='NN'])

# Print the results
```

```python
df.head()
```

```python
df['clean_text']=df['clean_text'].str.lower()
```

```python
diseases=['asthma','autism','dengue','fever','flu','nausea','anxiety','sprain','stroke','obesity']
```

```python
tw=[]
for i in df['clean_text']:
    tw.append(i)
```

```python
tag=[]
flag=False
for i in range(0,len(tw)):
    for j in range(0,len(diseases)):
        if(tw[i].find(diseases[j])>0):
            tag.append(diseases[j])
            flag=True
            break
    if(flag != True):
        tag.append('NIL')
    flag=False
```

```python
len(tag)
```

```python
df['tags']=tag
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
def find_matching_diseases(row):
    words = []
    for word in diseases:
        if word in row:
            words.append(word)
    return words
```

```python
df['matching_words'] = df['nouns'].apply(lambda x: find_matching_diseases(x))
```
