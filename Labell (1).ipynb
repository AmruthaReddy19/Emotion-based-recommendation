```python
df=df.drop(['Unnamed: 0','id','Time'],axis=1)
```

```python
diseas=['asthma','flu','fever','nausea','Nausea','Fever','Flu','Asthma','anxiety','sinus','Sinus','Anxiety','stroke','Stroke',
       'dengue','Dengue','obesity','Obesity','indigestion','Indigestion']
```

```python
newdis=[]
for i in range(0,len(diseas)):
    s=diseas[i]
    s=s.lower()
    newdis.append(s)
for i in newdis:
    diseas.append(i)
```

```python
diseas
```

```python
df
```

```python
tw=[]
for i in df['Tweet']:
    tw.append(i)
            
```

```python
tw
```

```python
health=[]
tag=[]
flag=False
for i in range(0,len(tw)):
    for j in range(0,len(diseas)):
        if(tw[i].find(diseas[j])>0):
            health.append(1)
            tag.append(diseas[j])
            flag=True
            break    
    if(flag!=True):
        tag.append("NIL")
        health.append(0);
    flag=False    
```

```python
len(tag)
```

```python

df["Tags"]=tag
```
