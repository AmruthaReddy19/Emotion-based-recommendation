```python
diseases
```

```python
tw=[]
for i in df['Tweet']:
    tw.append(i)
```

```python
len(tw)
```

```python
tag=[]
flag=False
for i in range(0,len(tw)):
    for j in range(0,len(diseases)):
        if(tw[i].find(diseases[j])>0):
            tag.append(diseases[j])
            flag=True
            break
    if(flag != True):
        tag.append('NIL')
    flag=False
```

```python
len(tag)
```

```python
df['tags']=tag
```
