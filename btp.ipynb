```python
import pandas as pd
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import *
stemmer = PorterStemmer()
```

```python
sample=pd.read_csv('bigtweet_file002.csv')
```

```python
sample.head()
```

```python
positive=pd.read_csv('positive-words.txt',header=None,encoding = "ISO-8859-1")
positive.columns=['words']
negative=pd.read_csv('negative-words.txt',header=None,encoding = "ISO-8859-1")
negative.columns=['words']
stop_words=stopwords.words('english')
```

```python
positive
```

```python
stop_words
```

```python
#preprocess function 
stem_words=lambda x: stemmer.stem(x)
break_into_words = lambda x : re.findall("[a-zA-Z0-9]+", x)
skip_stop_words = lambda x: [w for w in x if w not in list(stop_words)]
skip_numeric = lambda x : [w for w in x if not w.isnumeric()]
```

```python
#apply pre-processing steps
def lowercase(a):
    a = a.lower()
    return a
def break_into_words(x):
    return x.split()
def skip_stop_words(y):
    for i in y:
        if i in list(stop_words):
            y.remove(i)
    return y
def skip_numeric(y):
    for i in y:
        if i.isnumeric()==True:
            y.remove(i)
    return y
processed_text=sample['content'].map(lowercase)\
        .map(break_into_words)\
        .map(skip_stop_words)\
        .map(skip_numeric)
sample['processed']=processed_text
```

```python
sample
```

```python
#count number of positive words
positive_words_count=processed_text.map(lambda x: len([w for w in x if w in list(positive['words'])]))
#count number of negative words
negative_words_count=processed_text.map(lambda x: len([w for w in x if w in list(negative['words'])]))
sample['positive']=positive_words_count
sample['negative']=negative_words_count
#Finally, Calculate the Sentiment Score:
```

```python
sample
```

```python
l=len(sample['processed'])
sample['normalized_sentiment_score']=(sample['positive']-sample['negative'])/l
```
