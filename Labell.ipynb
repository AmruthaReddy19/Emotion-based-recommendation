```python
tw=[]
for i in df['Tweet']:
    tw.append(i)
            
```

```python
tw
```

```python
health=[]
tag=[]
flag=False
for i in range(0,len(tw)):
    for j in range(0,len(diseas)):
        if(tw[i].find(diseas[j])>0):
            health.append(1)
            tag.append(diseas[j])
            flag=True
            break    
    if(flag!=True):
        tag.append("NIL")
        health.append(0);
    flag=False    
```

```python
len(tag)
```

```python
df["Flag"]=health
df["Tags"]=tag
```
